```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips= sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.displot(tips['total_bill'], kde = True)
```

```python
#joint plot helps combine two different data

sns.jointplot(x='total_bill', y = 'tip',  data = tips , kind = 'kde')
```

```python
sns.pairplot(tips) #quickly visualize data
  # by adding a hew argument which is used for categroial column

```

```python
# by adding a hue argument which is used for categroial column

sns.pairplot(tips, hue = 'sex', palette='icefire')

# we can add pallete for differenet colors
```

```python
# for every single distribution it draws a point


sns.rugplot(tips['total_bill'])
```
